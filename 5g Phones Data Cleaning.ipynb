```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_json('5gPhones.json')
```

```python
df.head(10)
```

```python
df2 = df.drop(['processor'],axis=1)
```

```python
df2.describe()
```

```python
battery = df['battery']
```

```python
battery_processed=[]
ram_processed=[]
rom_processed=[]
for bat in battery:
    s=""
    for x in bat:
        if(x.isdigit()):
            s+=x
    battery_processed.append(int(s))
for ram in df['ram']:
    try:
        s=""
        for x in ram:
            if(x.isdigit()):
                s+=x
        ram_processed.append(int(s))
    except:
        ram_processed.append(None)
for rom in df['rom']:
    s=""
    for x in rom:
        if(x.isdigit()):
            s+=x
    rom_processed.append(int(s))
```

```python
df2['processed_battery'] = battery_processed
df2['ram_processed'] = ram_processed
df2['rom_processed'] = rom_processed
```

```python
df2.sample(5)
```

```python
df3 = df2[df2['processed_battery'] >= 4000]
```

```python
df3 = df3[df3['rom_processed']>=32]
df3 = df3[df3['ram_processed']>=2]
```

```python
backcam = []
for cam in df3['back_camera']:
    s=""
    for j in cam.split()[0]:
        if(j.isdigit()):
            s+=j
    backcam.append(int(s))
```

```python
df3['back_camera_processed'] = backcam
```

```python
frontcam = []
for cam in df3['front_camera']:
    
    try:
        s=""
        for j in cam.split()[0]:
            if(j.isdigit()):
                s+=j
        frontcam.append(int(s))
    except:
        frontcam.append(None)
```

```python
len(frontcam), len(df3['front_camera'].to_list())
```

```python
df3['front_camera_processed'] = frontcam
```

```python
screen=[]
for scr in df3['screen']:
    
    screen.append(float(scr.split()[0]))
```

```python
df3['screen_processed'] = screen
```

```python
brand = []
for n in df3['name']:
    s=n.split()[0]
    brand.append(s)
```

```python
df3['brand'] = brand
```

```python
price=[]
for p in df3['price']:
    
    s = re.sub(r"[^a-zA-Z0-9 ]", "", p)
    price.append(int(s))
```

```python
df3['price_processed'] = price
```

```python
final_df = df3.drop(['ram','rom','screen','battery','front_camera','back_camera','price'],axis=1)
```
